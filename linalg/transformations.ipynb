```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
origin = np.array(np.zeros((2,2)))
```

Common linear transformations:

```python
A = np.array([[2, 1], [-5, 3]])
plt.figure(figsize=(11,7))
plt.quiver(*origin, A[:,0], A[:,1], color=['r','b'], scale=20)
plt.show()
```

```python
clockwise = np.array([[0, 1], [-1, 0]]) #90 degrees
counter_clockwise = np.array([[0, -1], [1, 0]])
(np.linalg.inv(clockwise) == counter_clockwise).ravel().all()
```

```python
B = np.matmul(A, clockwise) #This rotation will be opposite, as we compute the inverse
plt.figure(figsize=(11,7))
plt.quiver(*origin, A[:,0], A[:,1], color=['r','b'], scale=20, pivot='tail')
plt.quiver(*origin, B[:,0], B[:,1], color=['g','y',], scale=20, pivot='tail')
plt.show()
```

```python
right_shear = [[1, 1], [0, 1]]
left_shear = [[1, -1], [0, 1]]
```

```python
C = np.matmul(A, right_shear)
plt.figure(figsize=(11,7))
plt.quiver(*origin, A[:,0], A[:,1], color=['r','b'], scale=20, pivot='tail')
plt.quiver(*origin, C[:,0], C[:,1], color=['g','y',], scale=20, pivot='tail')
plt.show()
```

```python
counter_clockwise_45 = np.power(2, 1/2) / 2 * np.array([[1, -1], [1, 1]])
```

To rotate $45^{\circ}$about the origin, we apply the matrix: 
$\\ \begin{vmatrix}
\frac{\sqrt{2}}{2} & \frac{-\sqrt{2}}{2} \\
\frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2}
\end{vmatrix} \\ $

Note, $\frac{\sqrt{2}}{2}=\cos45^{\circ}$, so for counter-clockwise this is the same as : $\\ \begin{vmatrix}
\cos \phi & -\sin \phi \\
\sin \phi & \cos \phi
\end{vmatrix}  =  \begin{vmatrix}
\cos45^{\circ} & -\sin45^{\circ} \\
\sin45^{\circ} & \cos45^{\circ}
\end{vmatrix} \\ $ 

```python
D = np.matmul(A, counter_clockwise_45)
plt.figure(figsize=(11,7))
plt.quiver(*origin, A[:,0], A[:,1], color=['r','b'], scale=20, pivot='tail')
plt.quiver(*origin, D[:,0], D[:,1], color=['g','y',], scale=20, pivot='tail')
plt.show()
```
